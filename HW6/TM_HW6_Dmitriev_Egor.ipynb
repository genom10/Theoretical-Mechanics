System is perfectly symmetrical so we can examine only half of it. The other half will behave exactly the same.

0. $E_0 = \frac{mgh}{2}$ - Syslem is stationaty. Only potential energy

### Subtask 1

$
\frac{mgh}{2} = \frac{mv_D^2}{2} + \frac{J\omega^2}{2}\\
J = mp^2;\ \omega = \frac{v_D}{l}\\
mgh = mv_D^2 + mp^2\frac{v_D^2}{l^2}\\
gh = v_D^2 + p^2\frac{v_D^2}{l^2}\\
gh = v_D^2(1 + \frac{p^2}{l^2})\\
v_D = \sqrt{\frac{ghl^2}{l^2+p^2}}\\
v_D = l\sqrt{\frac{gh}{l^2+p^2}}\\
v_B = 2v_D = 2l\sqrt{\frac{gh}{l^2+p^2}}
$
